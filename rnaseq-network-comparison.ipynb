```python
def get_path_name(name_db, input_file, cutoff=0.01):
    for n, line in enumerate(open(input_file)):
        if n == 0:
            continue
        items = line.strip().split('\t')
        term = items[0]
        name = items[-1]
        if float(items[5]) > cutoff:
            continue
        else:
            name_db[term] = name
```

```python
def get_evalue(input_file, cutoff=0.01):
    _dict = {}
    for n, line in enumerate(open(input_file)):
        if n == 0:
            continue
        items = line.split('\t')
        term = items[0]
        if float(items[5]) > cutoff:
            continue
        evalue = -1.0 * log10(float(items[1]))
        _dict[term] = evalue
    return _dict

def get_ngenes(input_file, cutoff=0.01):
    _dict = {}
    for n, line in enumerate(open(input_file)):
        if n == 0:
            continue
        items = line.split('\t')
        term = items[0]
        ngenes = float(items[-1])
        allgenes = float(items[4])
        if float(items[5]) > cutoff:
            continue
        _dict[term] = (ngenes/allgenes)*100.0
    return _dict
```

```python
cufflinks = get_evalue('human/cufflinks/line7u_vs_i.cuffref.degenes.KEGG.txt', 0.05)
gimme = get_evalue('human/gimme/line7u_vs_i.cuffref.degenes.KEGG.txt', 0.05)
ensbl = get_evalue('human/ensembl/line7u_vs_i.degenes.KEGG.txt', 0.05)
assm = get_evalue('human/assembly/line7u_vs_i.cuffref.degenes.KEGG.txt', 0.05)
```

```python
def get_gene_id(infile):
    genes = set()
    for n, line in enumerate(open(infile)):
        if n == 0:
            continue
        gene_name = line.strip().split('\t')[1]
        genes.add(gene_name)
    return n, genes
```

```python
gimme_n_genes, gimme_genes = get_gene_id('gallus/gimme/line7u_vs_i.cuffref.tophits.nucl.txt')
assm_n_genes, assm_genes = get_gene_id('gallus/assembly/line7u_vs_i.degenes.tophits.nucl.txt')
cufflinks_n_genes, cufflinks_genes = get_gene_id('gallus/cufflinks/line7u_vs_i.cuffref.tophits.nucl.txt')

ensbl_genes = set()
for n, line in enumerate(open('gallus/ensembl/line7u_vs_i.degenes.fdr.05')):
    if n == 0:
        continue
    gene = line.strip().split()[0].replace('"', '')
    ensbl_genes.add(gene)

ensbl_n_genes = n
```

```python
def get_kegg_genes(infile):
    genes = set()
    for n,line in enumerate(open(infile)):
        if n == 0:
            continue
        genes.add(line.split()[0])
    return genes
```
