```python
def get_path_id(db, input_file, cutoff=0.01):
    for n, line in enumerate(open(input_file)):
        if n == 0:
            continue
        items = line.strip().split('\t')
        cat = items[0]
        if float(items[5]) > cutoff:
            continue
        else:
            db.add('%0.5d' % int(cat))
```

```python
def get_path_name(*inputfiles):
    path_names = {}
    for f in inputfiles:
        with open(f) as fp:
            reader = csv.reader(fp, dialect='excel-tab')
            for item in reader:
                path_names[item[-1]] = item[0]
    return path_names
```

```python
def get_evalue(input_file, cutoff=0.01):
    _dict = {}
    reader = csv.DictReader(open(input_file), dialect='excel-tab')
    for row in reader:
        term = '%0.5d' % int(row['category'])
        if float(row['padjust']) > cutoff:
            continue
        evalue = -1.0 * log10(float(row['padjust']))
        _dict[term] = evalue
    return _dict
```

```python
def get_gene_id(infile):
    genes = set()
    ensbls = set()
    for n, line in enumerate(open(infile)):
        if n == 0:
            continue
        items = line.strip().split('\t')
        gene_name = items[0]
        ensbl_id = items[1]
        if ensbl_id == 'NA':
            continue
        else:
            genes.add(gene_name)
            ensbls.add(ensbl_id)
    return n, genes, ensbls
```

```python
def get_kegg_genes(infile):
    genes = set()
    for n,line in enumerate(open(infile)):
        if n == 0:
            continue
        genes.add(line.split()[0])
    return genes
```

```python
from collections import namedtuple, defaultdict

Count = namedtuple('Count', ['id', 'eff_length', 'est_count'])
```

```python
def get_read_counts(filename, annotfile):
    if annotfile:
        annots = {}
        for line in open(annotfile):
            geneid, ensembl = line.strip().split('\t')
            annots[geneid] = ensembl

    counts = {}
    for line in open(filename):
        if line.startswith('gene_id'):
            continue
        items = line.strip().split('\t')
        gene_id = items[0]
        eff_length = float(items[3])
        est_count = float(items[4])
        if annotfile != '':
            if gene_id in annots:
                counts[annots[gene_id]] = Count(gene_id, eff_length, est_count)
        else:
            counts[gene_id] = Count(gene_id, eff_length, est_count)
    return counts
```

```python
# estimated counts from the control group
ensbl_counts = get_read_counts('line7u-paired-rsem.genes.results.ensembl', '')
combined_counts = get_read_counts('line7u-paired-rsem.genes.results.gimme', 'line7u_vs_i.gimme.degenes.fdr.05.tophits.gga')
cuff_counts = get_read_counts('line7u-paired-rsem.genes.results.cufflinks', 'line7u_vs_i.cufflinks.degenes.fdr.05.tophits.gga')
assm_counts = get_read_counts('line7u-paired-rsem.genes.results.assembly', 'line7u_vs_i.assembly.degenes.fdr.05.tophits.gga')
```

```python
# estimated counts from the infected group
ensbl_counts_inf = get_read_counts('line7i-paired-rsem.genes.results.ensembl', '')
combined_counts_inf = get_read_counts('line7i-paired-rsem.genes.results.gimme',
                                          'line7u_vs_i.gimme.degenes.fdr.05.tophits.gga')
```

```python
common_genes = set(ensbl_counts).intersection(set(combined_counts))
common_genes.intersection_update(set(cuff_counts))
common_genes.intersection_update(set(assm_counts))
```

```python
len(common_genes)
```

```python
el = [] # ensembl gene lengths
cl = [] # gimme gene lengths
assml = [] # assembly gene lengths
cuffl = [] # cufflinks gene lengths
ec = [] # ensembl counts control
cc = [] # gimme counts control
ec_inf = [] # ensembl counts infected
cc_inf = [] # gimme counts infected

for gene_id in common_genes:
    el.append(ensbl_counts[gene_id].eff_length)
    cl.append(combined_counts[gene_id].eff_length)
    assml.append(assm_counts[gene_id].eff_length)
    cuffl.append(cuff_counts[gene_id].eff_length)
    
    ec.append(ensbl_counts[gene_id].est_count)
    cc.append(combined_counts[gene_id].est_count)
    ec_inf.append(ensbl_counts_inf[gene_id].est_count)
    cc_inf.append(combined_counts_inf[gene_id].est_count)
```

```python
norm_ec = [1000 * (ec[i] / el[i]) for i in range(len(ec))]
norm_cc = [1000 * (cc[i] / cl[i]) for i in range(len(cc))]
norm_ec_inf = [1000 * (ec_inf[i] / el[i]) for i in range(len(ec_inf))]
norm_cc_inf = [1000 * (cc_inf[i] / cl[i]) for i in range(len(cc_inf))]
```

```python
from scipy.stats import scoreatpercentile
```

```python
fig, ax = plt.subplots()
ax.boxplot([el, assml, cuffl, cl], sym='', )
ax.set_ylim(-1000, 10000)
#ax.set_xticks([1, 2, 3, 4], ['Ensembl', 'Assembly', 'Cufflinks', 'Merged'])
ax.set_ylabel('Effective gene size (bp)', size=16)
ax.set_xticklabels(['Ensembl', 'Assembly', 'Cufflinks', 'Merged'], rotation=85, fontsize=16)
ax.tick_params(axis='y', labelsize=14)

el_q1 = scoreatpercentile(el,25)
el_q3 = scoreatpercentile(el,75)
el_iqr = el_q3-el_q1

cl_q1 = scoreatpercentile(cl,25)
cl_q3 = scoreatpercentile(cl,75)
cl_iqr = cl_q3-cl_q1

cuffl_q1 = scoreatpercentile(cuffl,25)
cuffl_q3 = scoreatpercentile(cuffl,75)
cuffl_iqr = cuffl_q3-cuffl_q1

assml_q1 = scoreatpercentile(assml,25)
assml_q3 = scoreatpercentile(assml,75)
assml_iqr = assml_q3-assml_q1

savefig('gene_length.pdf', bbox_inches='tight', pad_inches=0.1)
print 'Ensembl: Median %.2f IQR %.2f, number = %d' % (median(el), el_iqr, len(el))
print 'Assembly: Median %.2f IQR %.2f, number = %d' % (median(assml), assml_iqr, len(assml))
print 'Cufflinks: Median %.2f IQR %.2f, number = %d' % (median(cuffl), cuffl_iqr, len(cuffl))
print 'Combined: Median %.2f IQR %.2f, number = %d' % (median(cl), cl_iqr, len(cl))
```
